```python
dt=pd.read_excel('ps.xlsx')
print(dt.shape)
print(dt.describe())
print(dt.head(2))
print(dt.tail(2))
```

```python
simpleRandomSample=dt.sample(n=4).sort_values(by='Id')
mean_s=round(simpleRandomSample['SalePrice'].mean(),3)
print("Mean:",mean_s)
print(simpleRandomSample)
```

```python
index=np.arange(0,len(dt),step=3)
systematicSample=dt.iloc[index]
mean_sy=round(systematicSample['SalePrice'].mean(),3)
print("Mean:",mean_sy)
print(systematicSample)
```

```python
n=13
dt['cluster_id']= np.repeat([range(1,n+1)],len(dt)/n)
index= []
for i in range(0,len(dt)):
  if dt['cluster_id'].iloc[i]%2==0:
    index.append(i)
clusterSample=dt.iloc[index]
mean_c=round(clusterSample['SalePrice'].mean(),3)
print("Mean:",mean_c)
print(clusterSample)
```

```python
n=2
dt['cluster_id']= np.tile(np.arange(1, n+1), len(dt) // n + 1)[:len(dt)]
index= []
for i in range(0,len(dt)):
  if dt['cluster_id'].iloc[i]%2==0:
    index.append(i)
clusterSample=dt.iloc[index]
mean_c=round(clusterSample['SalePrice'].mean(),3)
print("Mean:",mean_c)
print(clusterSample)
```

```python
dt['strata']=np.tile(np.arange(1, n+1), len(dt) // n + 1)[:len(dt)]
index=[]
for i in range(0,len(dt)):
    index.append(i)
strataSample=dt.iloc[index]
strataSample
```
